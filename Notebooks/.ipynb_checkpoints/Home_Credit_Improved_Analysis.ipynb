# Home Credit Default Risk: Improved Analysis
This notebook uses the modularized codebase to:
1. Load all datasets including previous applications, bureau balances, POS cash balances, credit card balances, and installment payments.
2. Inject domain specific features and handle anomalies.
3. Train three LightGBM models exploring different class imbalance strategies:
   - Baseline with class_weight='balanced'
   - SMOTE Oversampling
   - Random Undersampling
4. Compare model performances.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import gc

# Import modular helper scripts
import data_processing
import run_pipeline
import visualization_utils

# Display settings
pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings('ignore')
```

```python
print("Starting Data Processing via data_processing.py...")
# debug=False will load the full dataset. Set to True for a quick test.
df = data_processing.main(debug=True)

# Separate into train and test
train_df = df[df['TARGET'].notnull()].copy()
test_df = df[df['TARGET'].isnull()].copy()
print(f"Train Shape: {train_df.shape}")
print(f"Test Shape: {test_df.shape}")
```

```python
print("Training Baseline Model (HistGradientBoosting / LightGBM equivalence)...")
from sklearn.ensemble import HistGradientBoostingClassifier
from imblearn.pipeline import Pipeline

baseline_pipeline = Pipeline([
    ('classifier', HistGradientBoostingClassifier(
        max_iter=500,
        learning_rate=0.05,
        max_leaf_nodes=31,
        class_weight='balanced',
        max_depth=8,
        l2_regularization=0.1,
        early_stopping=True,
        random_state=50
    ))
])

oof_base, preds_base = run_pipeline.train_model(train_df, test_df, pipeline=baseline_pipeline)
from sklearn.metrics import roc_auc_score
auc_base = roc_auc_score(train_df['TARGET'], oof_base)
print(f"Baseline AUC: {auc_base:.4f}")
```

```python
print("Training SMOTE Model...")
from imblearn.over_sampling import SMOTE
from sklearn.impute import SimpleImputer

smote_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('sampler', SMOTE(random_state=50)),
    ('classifier', HistGradientBoostingClassifier(
        max_iter=500,
        learning_rate=0.05,
        max_leaf_nodes=31,
        max_depth=8,
        l2_regularization=0.1,
        early_stopping=True,
        random_state=50
    ))
])

oof_smote, preds_smote = run_pipeline.train_model(train_df, test_df, pipeline=smote_pipeline)
auc_smote = roc_auc_score(train_df['TARGET'], oof_smote)
print(f"SMOTE AUC: {auc_smote:.4f}")
```

```python
print("Training Random Undersampling Model...")
from imblearn.under_sampling import RandomUnderSampler

under_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('sampler', RandomUnderSampler(random_state=50)),
    ('classifier', HistGradientBoostingClassifier(
        max_iter=500,
        learning_rate=0.05,
        max_leaf_nodes=31,
        max_depth=8,
        l2_regularization=0.1,
        early_stopping=True,
        random_state=50
    ))
])

oof_under, preds_under = run_pipeline.train_model(train_df, test_df, pipeline=under_pipeline)
auc_under = roc_auc_score(train_df['TARGET'], oof_under)
print(f"Undersampling AUC: {auc_under:.4f}")
```

```python
print("--- Model Performance Comparison ---")
print(f"Baseline (class_weight): {auc_base:.4f}")
print(f"SMOTE Oversampling:       {auc_smote:.4f}")
print(f"Random Undersampling:     {auc_under:.4f}")

# Plot combined ROC Curves
from sklearn.metrics import roc_curve
plt.figure(figsize=(8, 6))

fpr_b, tpr_b, _ = roc_curve(train_df['TARGET'], oof_base)
fpr_s, tpr_s, _ = roc_curve(train_df['TARGET'], oof_smote)
fpr_u, tpr_u, _ = roc_curve(train_df['TARGET'], oof_under)

plt.plot(fpr_b, tpr_b, label=f'Baseline (AUC = {auc_base:.4f})')
plt.plot(fpr_s, tpr_s, label=f'SMOTE (AUC = {auc_smote:.4f})')
plt.plot(fpr_u, tpr_u, label=f'Undersampling (AUC = {auc_under:.4f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves Comparison')
plt.legend(loc='lower right')
plt.show()
```
